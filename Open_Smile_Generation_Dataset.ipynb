```python
list_of_filenames=os.listdir("/home/bigfloppa/Рабочий стол/Datasets/SAVEE")
```

```python
feeling_list=[]
for item in list_of_filenames:
    if item[2:4]=='_a':
        feeling_list.append('-')
    elif item[2:4]=='_d':
        feeling_list.append('-')
    elif item[2:4]=='_f':
        feeling_list.append('-')
    elif item[2:4]=='_h':
        feeling_list.append('+')
    elif item[2:4]=='_n':
        feeling_list.append('n')
    elif item[3:5]=='sa':
        feeling_list.append('-')
    elif item[3:5]=='su':
        feeling_list.append('n')
```

```python
labels = pd.DataFrame(feeling_list)
labels=labels.rename(columns={0:"class"})
labels
```

```python
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv02,
    feature_level=opensmile.FeatureLevel.Functionals,
)
```

```python
Data=pd.DataFrame()

for index,y in tqdm(enumerate(list_of_filenames)):
    p='/home/bigfloppa/Рабочий стол/Datasets/SAVEE/'+y
    signal, sampling_rate = audiofile.read(
    p,
    duration=3,
    always_2d=True,)
    
    row=smile.process_signal(
    signal,
    sampling_rate)
    
    Data=pd.concat([Data,row])
    
      
```

```python
Data
```

```python
Data=Data.reset_index()
```

```python
Data=Data.drop(['start','end'],axis=1)
Data
```

```python
Data=pd.concat([Data,labels],axis=1)
Data
```
